### Data faltante

```python
df[num_vars].isna().sum()
```

Al transformar la data categórica, se asignó un código para los datos faltantes.

En el caso de data numérica debemos hacer un tratamiento especial a la data faltante.


El procedimiento que realizaremos es el siguiente:
1. Obtenemos un diccionario con las medianas de cada variable.
2. Creamos una columna que indique si hay data faltante.
3. Reemplazamos los datos faltantes con la mediana obtenida.

```python
# 1. Obtenemos la mediana de la columna y la almacenamos en un diccionario.
na_dict = df[num_vars].median().to_dict()
na_dict
```

```python
# 2. Creamos una columna que indique si hay data faltante.
df['Age_na'] = df['Age'].isna()
df.head()
```

```python
# 3. Reemplazamos los datos faltantes con la mediana obtenida.
df.Age.fillna(na_dict['Age'], inplace=True)
df[num_vars].isna().sum()
```

### TRABAJO | RESOLUCIÓN

El objetivo del presente trabajo es poder obtener una mayor seguridad en la predicción hecha por el árbol de decisión. En esa medida, se optará por desarrollar un KFoldCV y una búsqueda Grid. La primera servirá para poder dividir la data general **df_train** en TRAIN/VALIDATION. La segunda servirá para poder buscar los mejores hiperparámetros para el árbol de decisión. Estos hiperparámetros serán:

- '_max_depth_': máxima densidad el árbol de decisión
- '_min_sample_leaf_': muestras mínimas requeridas para la const. de un nodo/hoja
- '_min_sample_split_': muestras mínimas requeridas para dividir un nodo interno.



```python
#Importamos KFold y GridSearch
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV

#definimos número de folds
kfolds = KFold(n_splits=5, shuffle=True)
```

```python
#creamos diccionario vacío para albergar modelos tuneados
dict_models_score_fold = {}

#iteración x Fold
for train_idx, val_idx in kfolds.split(x):
  #división entre x_train e y_train x Fold coord
  x_train, y_train = x.iloc[train_idx], y.iloc[train_idx]
  #división entre x_val e y_val x Fold coord
  x_val, y_val = x.iloc[val_idx], y.iloc[val_idx]
      
  #parámetros a tunear: 
  #   máxima profundidad del árbol
  #   mínimo número de muestras para una hoja  
  #   mínimo número de splits para conf. nodo 
  tuned_params = {
          'max_depth':list(
              range(1,10)
              ),
          'min_samples_leaf': list(
              range(1,10)
              ),
          'min_samples_split':list(
              np.arange(0.1, 1.0, 0.1)
              )
          }

  #definición de modelo vacío
  m = DecisionTreeClassifier()

  #GridSearch simple
  model = GridSearchCV(
                  estimator=m,
                  param_grid=tuned_params,
                  cv=2
                  )
  #fiteamos el modelo      
  model.fit(x_train, y_train)

  #extraemos el modelo con el mejor estimador
  modelTemp = model.best_estimator_ 

  #corroboramos score OOF
  dict_models_score_fold[modelTemp] = modelTemp.score(x_val,y_val)
```

```python
#seleccionamos el modelo con mayor score OOF
selected_model = max(
      dict_models_score_fold,
      key=dict_models_score_fold.get
      )
```

```python
#modelo seleccionado
selected_model
```

```python
#score OOF en ValidationSet del modelo
dict_models_score_fold[selected_model]
```

```python
#predicción ejecutada
selected_model.predict(df_test)
```

## Rpta.

De manera inicial, pudimos observar que el conjunto de datos **df_test** no posee labels de validación. Esto significa que debemos aprovechar el conjunto **df_train** al máximo. Por tanto, se optó por dividir dicha data en TRAIN/VALIDATION. Esto sirvió para, en primera instancia, poder definir los hiperparámetros más ideales del modelo con base a la data TRAIN. Posteriormente, probar este modelo tuneado por c/fold con la data de validación (o VALIDATION SET). De ese modo, el modelo elige los mejores parámetros con base al conjunto de entrenamiento. Sin embargo, solo aquel modelo con mayor score en la data de validación (no utilizada para la elección de hiperparámetros) será el elegido. Esto nos permitirá poder confiar en un nivel de score OOF (out-of-sample) lo más alto posible.
